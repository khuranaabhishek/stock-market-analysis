```python
#CHATGPT

import matplotlib.pyplot as plt
import pandas as pd

# Define the number of subplots (assuming you want 10 rows and 5 columns)
rows = 10
cols = 5

# Create the figure and subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 20))

# Flatten axes array for easier access
axes = axes.flatten()

# Loop through each symbol to plot on a different subplot
for i, s in enumerate(symbols):
    df = pd.read_csv(f'{s} price details.csv')
    x = df.index
    y = df['Close']
    
    # Use the corresponding subplot axis (axes[i])
    ax = axes[i]
    ax.plot(x, y, label=s, color="black")
    ax.set_title(s)
    ax.set_xlabel("Date")
    ax.set_ylabel("Closing prices")
    ax.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
